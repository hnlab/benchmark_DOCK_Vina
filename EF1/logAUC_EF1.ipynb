```python
#2022-5-21, all targets in DUD-E dataset.
# EF1 after consider classification of target.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

def add_median_labels(ax):
    lines = ax.get_lines()
    # determine number of lines per box (this varies with/without fliers)
    boxes = [c for c in ax.get_children() if type(c).__name__ == 'PathPatch']
    lines_per_box = int(len(lines) / len(boxes))
    # iterate over median lines
    for median in lines[4:len(lines):lines_per_box]:
        # display median value at center of median line
        x, y = (data.mean() for data in median.get_data())
        # choose value depending on horizontal or vertical plot orientation
        value = x if (median.get_xdata()[1]-median.get_xdata()[0]) == 0 else y
        text = ax.text(x, y, f'{value:.1f}', ha='center', va='center',
                       fontweight='normal', color='white',fontsize='x-small')
        # create median-colored border around white text for contrast
        text.set_path_effects([
            path_effects.Stroke(linewidth=3, foreground=median.get_color()),
            path_effects.Normal(),
        ])
f, ax = plt.subplots(figsize=(6, 4))
data = pd.read_csv("/home/mxu02/dock37-test/test_dude/script/EF1/dock_vina_logAUC_ef1.csv", sep=",",header=0)

ax = sns.boxplot(data=data, x="Classification", y="EF1",hue="Method")
# add_median_labels(ax.axes)
ax.set_xticklabels(data.Classification, rotation=45,fontsize=8)

loc = 'best'
ax=plt.xlabel("Target family", size=10)
ax=plt.ylabel('${EF_1}$', size=10)
ax=plt.legend(loc=loc, fontsize='small')
# ax=plt.show()
plt.savefig("/home/mxu02/dock37-test/test_dude/script/EF1/target102_ef1.png", dpi=200, bbox_inches='tight')
```

```python
#EF1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

def add_median_labels(ax):
    lines = ax.get_lines()
    # determine number of lines per box (this varies with/without fliers)
    boxes = [c for c in ax.get_children() if type(c).__name__ == 'PathPatch']
    lines_per_box = int(len(lines) / len(boxes))
    # iterate over median lines
    for median in lines[4:len(lines):lines_per_box]:
        # display median value at center of median line
        x, y = (data.mean() for data in median.get_data())
        # choose value depending on horizontal or vertical plot orientation
        value = x if (median.get_xdata()[1]-median.get_xdata()[0]) == 0 else y
        text = ax.text(x, y, f'{value:.1f}', ha='center', va='center',
                       fontweight='normal', color='white',fontsize='large')
        # create median-colored border around white text for contrast
        text.set_path_effects([
            path_effects.Stroke(linewidth=3, foreground=median.get_color()),
            path_effects.Normal(),
        ])
f, ax = plt.subplots(figsize=(3, 4))
data = pd.read_csv("/home/mxu02/dock37-test/test_dude/script/EF1/dock_vina_logAUC_ef1.csv", sep=",",header=0)

ax = sns.boxplot(data=data, x="Method", y="EF1")
ax = sns.swarmplot(
    x="Method",
    y='EF1',
    data=data,
    dodge=True,
    linewidth=1.5,
    )
add_median_labels(ax.axes)
ax.set_xticklabels(data.Method, rotation=45,fontsize=8)

loc = 'best'
ax=plt.xlabel("Method", size=10)
ax=plt.ylabel('${EF_1}$', size=10)
# ax=plt.legend(loc=loc, fontsize='small')
# ax=plt.show()
plt.savefig("/home/mxu02/dock37-test/test_dude/script/EF1/ef1_102.png", dpi=200, bbox_inches='tight')
```
